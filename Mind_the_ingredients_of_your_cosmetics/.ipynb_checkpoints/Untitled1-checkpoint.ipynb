```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('datasets/cosmetics.csv', 
    dtype={
    'combination': 'bool',
    'dry': 'bool',
    'normal':'bool',
    'oily': 'bool',
    'sensitive': 'bool'
    },
    names=[
        'label',
        'brand', 
        'name', 
        'price', 
        'rank', 
        'ingredients', 
        'combination', 
        'dry', 
        'normal', 
        'oily', 
        'sensitive'], 
    header=0)
```

```python
df['id'] = df.index
```

```python

df = (df.loc[df['ingredients'] != 'No Info'])
```

```python
pattern = "^Visit.*"
filter = df['ingredients'].str.contains(pattern)
df = df[~filter]
```

```python
df['ingredients'] = df['ingredients'].str.replace(r'\*+', '', regex=True)
```

```python
df['brand'] = df['brand'].str.replace("KIEHL'S SINCE 1851", "KIEHL'S")
```

```python
labels = df['label'].unique().tolist()

brands = df['brand'].unique().tolist()

max_price = df['price'].max()

min_price = df['price'].min()
```

```python
df = (df.loc[df['name'] != '#NAME?'])
```

```python
df = (df.loc[df['ingredients'] != '#NAME?'])
```

```python
df['ingredients'] = df['ingredients'].str.replace(r'\(\+/-\):', '', regex=True)
```

```python
df['ingredients'] = df['ingredients'].str.replace(r'\(-\)-', '', regex=True)
```

```python
df['ingredients'] = df['ingredients'].str.replace(r'\+/-:', '', regex=True)
```

```python
df['ingredients'] = df['ingredients'].str.replace(r'\+', '', regex=True)
```

```python
df['ingredients'] = df['ingredients'].str.replace(r'-', '', regex=True)
```

```python
df['ingredients'] = df['ingredients'].str.replace(r'1,2Hexanediol,Spirulina Platensis Powder', '1,2Hexanediol, Spirulina Platensis Powder', regex=True)
```

```python
df['ingredients'] = df['ingredients'].str.replace(r'1?[,.]?2\s?Hexane?diol\s?', '1,2 Hexanediol', regex=True, flags=re.I)
```

```python
df['ingredients'] = df['ingredients'].str.replace('4% Glycolic Acid Complex: Helps to exfoliate dead surface skin cells and helps stimulate cellular turnover. Allantoin: Helps soothe and calm the skin. Aloe Vera: Helps soothe and hydrate the skin.', 
                                                  '4% Glycolic Acid Complex, Allantoin, Aloe Vera, ')
```

```python
df['ingredients'] = df['ingredients'].str.replace('4MSK: Helps remove melanin trapped in dark spots and ion force complex works to slough off the surface cells that contain melanin.', 
                                                 '4MSK')
```

```python
df.to_csv('datasets/cosmetic_filter.csv')
```
