```python
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
nobel = pd.read_csv('datasets/nobel.csv')

display(nobel.head())

print(nobel.info())

display(nobel['birth_country'].value_counts().head())
```

```python
nobel['birth_country'] = nobel['birth_country'].str.replace('United States of America', 'USA')
nobel['birth_country'] = nobel['birth_country'].str.replace('United Kingdom', 'U.K')
nobel['birth_country'] = nobel['birth_country'].str.replace(r'.*\(Poland\)', 'Poland', regex=True)
nobel['birth_country'] = nobel['birth_country'].str.replace(r'.*\(Germany\)', 'Germany', regex=True)
nobel['birth_country'] = nobel['birth_country'].str.replace(r'.*\(Russia\)', 'Russia', regex=True)
nobel['birth_country'] = nobel['birth_country'].str.replace(r'.*\(France\)', 'France', regex=True)
```

```python
countries = ['USA', 'U.K', 'Germany', 'France', 'Sweden', 'Russia', 'Japan', 'Poland']
def others_countries(country):
    if country in countries:
        return country
    return 'others'

nobel['country'] = nobel['birth_country'].apply(others_countries)
nobel['country'].unique()
```

```python
nobel['decade'] = (np.floor(nobel['year']/10) * 10).astype(int)
```

```python
dataframe = nobel.groupby(['country', 'category', 'decade']).count()
display(dataframe)

dataframe1 = dataframe[['sex']]
display(dataframe1)
```

```python
from bokeh.io import output_file, show, curdoc
from bokeh.models import ColumnDataSource, HoverTool, Select
from bokeh.plotting import figure
from bokeh.sampledata.commits import data
from bokeh.transform import jitter, factor_cmap
from bokeh.colors import HSL
from bokeh.layouts import row

#output_notebook()


categories = nobel['category'].unique().tolist()
countries = nobel['country'].unique().tolist()
print(countries)
organization_names = nobel['organization_name'].unique().tolist()


cmap = {
    "Chemistry"  : HSL(0, 0.5, 0.5).to_rgb().to_hex(),
    "Literature" : HSL(60, 0.5, 0.5).to_rgb().to_hex(),
    "Medicine"   : HSL(120, 0.5, 0.5).to_rgb().to_hex(),
    "Physics"    : HSL(180, 0.5, 0.5).to_rgb().to_hex(),
    "Peace"      : HSL(240, 0.5, 0.5).to_rgb().to_hex(),
    "Economics"  : HSL(300, 0.5, 0.5).to_rgb().to_hex(),
}


source = ColumnDataSource(nobel)

hover = HoverTool(tooltips=[('country', '@country')])

p = figure(plot_width=1600, plot_height=1600, x_range=organization_names, y_range=categories,
           title="Commits by Time of Day (US/Central) 2012—2016")

p.hex(x=jitter('organization_name', width=0.5, range=p.x_range), y=jitter('category', width=0.5, range=p.y_range), size=8, 
         source=source, alpha=0.8, color=factor_cmap('category', palette=list(cmap.values()), factors=list(cmap.keys())))

p.add_tools(hover)

p.x_range.range_padding = 0.3
p.y_range.range_padding = 0.3
#p.xaxis.ticker = list(range(1900, 2020, 10))
show(p)
```
